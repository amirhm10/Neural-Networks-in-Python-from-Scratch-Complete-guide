# Implementing Perceptron 3
![image](perceptron2.png)

```python
import numpy as np

inputs = np.array([[0,0], [0,1], [1,0], [1,1]])
inputs.shape
```

```python
outputs = np.array([0, 0, 0, 1])
outputs.shape
```

```python
weights = np.array([0.0, 0.0])
```

```python
learning_rate = 0.1
```

```python
def step_func(sum):
    if (sum>=1):
        return 1
    return 0
```

```python
def calculate_output(instance):
    s = np.dot(instance, weights)
    return step_func(s)
calculate_output(np.array([0,1]))
```

```python
def train():
    total_error = 1
    while (total_error != 0):
        total_error = 0
        for i in range(len(outputs)):
            prediction = calculate_output(inputs[i])
            error = abs(outputs[i]-prediction)
            total_error += error
            if error>0:
                for j in range(len(weights)):
                    weights[j]=weights[j]+learning_rate*inputs[i,j]*error
                    print("Weight updated: "+str(weights[j]))
        print("total error: "+str(total_error))
```

```python
train()
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
inputs = np.array([[18,2], [20,3], [21, 4],[35,15], [36,16], [38, 18]])
inputs = sc.fit_transform(inputs)
outputs = np.array([0,0,0,1,1,1])
outputs.shape
weights = np.array([0.0, 0.0])
learning_rate = 0.1
def step_func(sum):
    if (sum>=1):
        return 1
    return 0
def calculate_output(instance):
    s = np.dot(instance, weights)
    return step_func(s)
def train():
    total_error = 1
    while (total_error != 0):
        total_error = 0
        for i in range(len(outputs)):
            prediction = calculate_output(inputs[i])
            error = abs(outputs[i]-prediction)
            total_error += error
            if error>0:
                for j in range(len(weights)):
                    weights[j]=weights[j]+learning_rate*inputs[i,j]*error
                    print("Weight updated: "+str(weights[j]))
        print("total error: "+str(total_error))
```

```python
train()
```

```python
test_inputs = np.array([[17,5], [25,8], [45,10], [31,20]])
test_inputs = sc.transform(test_inputs)
for i in range(len(test_inputs)):
    d = calculate_output(test_inputs[i])
    print(d)
```
